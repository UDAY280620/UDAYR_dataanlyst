```python
df = pd.read_csv('covid_clinical_trials.csv')
```

```python
df.fillna({'Results First Posted':'Unknown'}, inplace=True)
```

```python
 print(df['Status'].value_counts())
```

```python
df['Status'].value_counts().plot(kind='bar', title='Status of Clinical Trials')
```

```python
print(df['Phases'].value_counts())
```

```python
df['Phases'].value_counts().plot(kind='bar', title='Distribution of Phases')
```

```python
print(df['Age'].value_counts())
```

```python
df['Age'].value_counts().plot(kind='bar', title='Age Group Distribution')
```

```python
status_phase = pd.crosstab(df['Status'], df['Phases'])
```

```python
 print(status_phase)
```

```python
status_phase.plot(kind='bar', stacked=True, title='Status vs. Phases')
```

```python
conditions_outcomes = df.groupby('Conditions')['Outcome Measures'].apply(lambda x: ', '.join([str(i) for i in x if isinstance(i, str)])).reset_index()
```

```python
print(conditions_outcomes)
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
```

```python
df['Primary Completion Date'] = pd.to_datetime(df['Primary Completion Date'], errors='coerce')
```

```python
df['Start Date'].dt.to_period('M').value_counts().sort_index().plot(kind= 'line', title='Trials Started Over Time')
```

```python
df.to_csv('cleaned_covid_clinical_trials.csv', index=False)
```

## Conclusion

##### The majority of trials are in the "Completed" phase

```python
Trial Status: A large proportion of trials are listed as “Completed,” reflecting the surge in research activity during the peak of the pandemic.
```

##### Most trials target adult populations

```python
Target Population: Most trials focus on adult participants, with fewer targeting children, elderly subgroups, or pregnant populations.
```

##### There's a steady increase in the number of trials over time

```python
Trial Timing: There’s often a steady rise in the number of trials initiated from early 2020 onward, peaking in mid-to-late 2021 before slowing in subsequent years.
```
